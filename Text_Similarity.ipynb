```python
import re
import string
```

```python
import spacy.cli
spacy.cli.download("pt_core_news_lg")
```

```python
nlp = spacy.load("pt_core_news_lg")
```

```python
# prompt: read rtf file and allocate its content into a string variable

arquivo = open("vagas_emprego.txt", "r",encoding="Latin-1")
vagas = arquivo.read()
vagas
```

```python
vagas = vagas.replace("\n", " ").strip()
vagas = vagas.replace("\+", " ").strip()
vagas = "".join([char for char in vagas if char not in string.punctuation])
vagas = re.sub(' +', ' ', vagas)
vagas = vagas.lower()
doc1 = nlp(vagas)
vagas
```

```python
doc1
```

```python
tokens1 = [token for token in doc1 if not token.is_stop]
```

```python
# prompt: convert my list of tokens to a list of strings

string_tokens = [str(token) for token in tokens1]
print(string_tokens, sep=",")
```

```python
# prompt: count of ocurrences of strings in string_tokens2
from collections import Counter
counts = Counter(string_tokens)

count_words = []
for word, count in counts.items():
    count_words.append((word, count))

print(count_words)

# Print the counts of each string in string_tokens2
#for string, count in counts.items():
#    print(f"{string}: {count}")
```

```python
# prompt: print count_words sort by "count" value

count_words.sort(key=lambda x: x[1], reverse=True)
print(count_words)
```
